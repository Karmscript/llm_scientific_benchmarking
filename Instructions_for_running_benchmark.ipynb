#**Instructions for running the designed benchmark**

#**Steps to run benchmark**

1. Having unzipped the folder in which this Instruction was found, open the notebook title **"Benchmark_for_Gemini.ipynb"**

2. Ensure you're using Google colab pro as Gemini 3 pro cannot be used with the free version of Google colab. Also prepare your API key.

3. Look for this line in the notebook:
   ```
   from google import genai

   from google.colab import userdata
   GEMINI_API_KEY= userdata.get('COLAB_GEM')

   # The client gets the API key from the environment variable `GEMINI_API_KEY`.
   client = genai.Client(api_key=GEMINI_API_KEY)
   ```
4. Replace 'COLAB_GEM' with the name of your valid api key that you have added to your notebook. Make sure your key can be accessed by all notebooks.

5. Click 'run all'. Then scroll down to the unit tests sections to view the results of each tests.

6. Go to last line in the notebook to view the overall evaluation results in JSON.

7. **If any API call fails during the course of running the notebook** due to server issues, simply **Run focused cells and all cells below**

**More on step 4**

1. Open the "Secrets" tab (the key icon 🔑) in the left sidebar of Google Colab.

2. Add a new secret with your preferred name

3. Paste your Gemini API Key into the Value field. Go to the Google AI studio using [https://aistudio.google.com/?project=gen-lang-client-0850758115], click the menu bar and click the "Get API key" button to get a new API key.

4. Toggle the "Notebook access" switch to On.
